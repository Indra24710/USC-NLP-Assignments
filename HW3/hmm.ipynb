```python
class HMM:
    def __init__(self):
        self.words={}
        self.postags={}
        self.prevpos={}
        self.transprob={}
        self.openclasstags=[]
    
    def split(self,word):
        j=len(word)-1
        while word[j]!='/':
            j-=1
        return word[:j],word[j+1:]
        
    def create_corpora(self,trainset):
        
        for sentence in trainset:
            sentence=sentence.decode("utf-8")
            sentence=sentence.rstrip()
            wordlist=sentence.split(" ")
            prev='start'
            if 'start' not in self.postags:
                self.postags['start']=[0,[]]
            self.postags['start'][0]+=1
            for word in wordlist:
                w,tag=self.split(word)
                if w not in self.words:
                    self.words[w]={}
                    self.words[w]['tcount']=0
                if tag not in self.words[w]:
                    self.words[w][tag]=0
                if tag not in self.postags:
                    self.postags[tag]=[0,[]]
                    
                self.words[w][tag]+=1
                self.words[w]['tcount']+=1
                self.postags[tag][0]+=1
                self.postags[tag][1].append(w)
                if prev:
                    if prev not in self.prevpos:
                        self.prevpos[prev]={}
                        self.transprob[prev]={}
                    if tag not in self.prevpos[prev]:
                        self.prevpos[prev][tag]=0
                        self.transprob[prev][tag]=0
                    self.transprob[prev][tag]=(self.prevpos[prev][tag]+1)/(self.postags[prev][0]+len(self.postags.keys()))
                    
                    self.prevpos[prev][tag]+=1
                    
                prev=tag
                
    
    def predict_tags(self,sentence):
        
        sentence=sentence.decode("utf-8")
        sentence=sentence.rstrip()
        sentence=sentence.split(" ")
        
        opentags=set(self.openclasstags)
        taglist=list(self.postags.keys())
        tag_ind={}
        for i in range(len(taglist)):
            tag_ind[taglist[i]]=i

        prev='start'
        word=sentence[0]
        i=0
        m_ind=[float("-inf"),None]
        viterbi_matrix={}
        viterbi_matrix[i]={}
        viterbi_matrix[i+1]={}
        if word not in self.words.keys():
            
            for tag in opentags:
                if tag not in self.transprob[prev].keys():
                    continue
                prob=self.transprob[prev][tag]
                viterbi_matrix[i][tag]=[prob,'start']
                if i!=len(sentence)-1:
                    for nextag in self.transprob[tag].keys():
                        if nextag not in viterbi_matrix[i+1]:
                            viterbi_matrix[i+1][nextag]=[float("-inf"),None]
                        if prob*self.transprob[tag][nextag]>viterbi_matrix[i+1][nextag][0]:
                            viterbi_matrix[i+1][nextag]=[prob*self.transprob[tag][nextag],tag]
                else:
                    if prob>m_ind[0]:
                        m_ind=[prob,tag]
        else:
            for tag in self.words[word].keys():
                if tag not in self.transprob[prev].keys():
                    continue
                prob=self.transprob[prev][tag]*(self.words[word][tag]/self.postags[tag][0])
                viterbi_matrix[i][tag]=[prob,'start']
                if i!=len(sentence)-1:
                    for nextag in self.transprob[tag].keys():
                        if nextag not in viterbi_matrix[i+1]:
                            viterbi_matrix[i+1][nextag]=[float("-inf"),None]
                        if prob*self.transprob[tag][nextag]>viterbi_matrix[i+1][nextag][0]:
                            viterbi_matrix[i+1][nextag]=[prob*self.transprob[tag][nextag],tag]
                else:
                    if prob>m_ind[0]:
                        m_ind=[prob,tag]

        
        
        for i in range(1,len(sentence)):
            word=sentence[i]
            viterbi_matrix[i+1]={}
            skipcount=0
            if word not in self.words.keys():
                for tag in viterbi_matrix[i].keys():
                    if tag not in opentags:
                        continue
                    prev=viterbi_matrix[i][tag][1]
                    prob=self.transprob[prev][tag]*viterbi_matrix[i][tag][0]
                    if i!=len(sentence)-1:
                        for nextag in self.transprob[tag].keys():
                            if nextag not in viterbi_matrix[i+1]:
                                viterbi_matrix[i+1][nextag]=[float("-inf"),None]
                            if prob*self.transprob[tag][nextag]>viterbi_matrix[i+1][nextag][0]:
                                viterbi_matrix[i+1][nextag]=[prob*self.transprob[tag][nextag],tag]
                    else:
                        if prob>m_ind[0]:
                            m_ind=[prob,tag]
            else:
                for tag in viterbi_matrix[i].keys():
                    if tag not in self.words[word].keys():
                        skipcount+=1

                        continue
                    prev=viterbi_matrix[i][tag][1]
                    prob=self.transprob[prev][tag]*(self.words[word][tag]/self.postags[tag][0])*viterbi_matrix[i][tag][0]
                    if i!=len(sentence)-1:
                        for nextag in self.transprob[tag].keys():
                            if nextag not in viterbi_matrix[i+1]:
                                viterbi_matrix[i+1][nextag]=[float("-inf"),None]
                            if prob*self.transprob[tag][nextag]>viterbi_matrix[i+1][nextag][0]:
                                viterbi_matrix[i+1][nextag]=[prob*self.transprob[tag][nextag],tag]
                    else:
                        if prob>m_ind[0]:
                            m_ind=[prob,tag]
            if skipcount==len(viterbi_matrix[i].keys()):
                for tag in viterbi_matrix[i].keys():
                    if tag not in opentags:
                        continue
                    prev=viterbi_matrix[i][tag][1]
                    prob=self.transprob[prev][tag]*viterbi_matrix[i][tag][0]
                    if i!=len(sentence)-1:
                        for nextag in self.transprob[tag].keys():
                            if nextag not in viterbi_matrix[i+1]:
                                viterbi_matrix[i+1][nextag]=[float("-inf"),None]
                            if prob*self.transprob[tag][nextag]>viterbi_matrix[i+1][nextag][0]:
                                viterbi_matrix[i+1][nextag]=[prob*self.transprob[tag][nextag],tag]
                    else:
                        if prob>m_ind[0]:
                            m_ind=[prob,tag]

                
        j=m_ind[1]
        outputs=[j]
        if len(sentence)>1:
#             print(j,sentence,viterbi_matrix[len(sentence)-3])
            nexj=viterbi_matrix[len(sentence)-1][j][1]
            for i in range(len(sentence)-2,-1,-1):
                outputs=[nexj]+outputs
                nexj=viterbi_matrix[i][nexj][1]

        s=[]
        for i in range(len(sentence)):
            s.append(sentence[i]+"/"+outputs[i])
                
        return " ".join(s)
```

```python
f=open('./hmm-training-data/ja_gsd_train_tagged.txt','rb')
train_data=f.readlines()
hmm_ita=HMM()
hmm_ita.create_corpora(train_data)
open_class_tags=[[i,len(set(hmm_ita.postags[i][1]))] for i in hmm_ita.postags.keys()]
open_class_tags=sorted(open_class_tags,key=lambda x:x[1])[::-1]
open_class_tags=[i[0] for i in open_class_tags]
hmm_ita.openclasstags=open_class_tags[:5]
pred_outputs=[]
f=open('./hmm-training-data/ja_gsd_dev_raw.txt','rb')
for i in f.readlines():
    o=hmm_ita.predict_tags(i)
    o=o.split(" ")
    pred_outputs+=o
act_outputs=[]
f=open('./hmm-training-data/ja_gsd_dev_tagged.txt','rb')
for i in f.readlines():
    sentence=i.decode("utf-8")
    sentence=sentence.rstrip()
    o=sentence.split(" ")
    act_outputs+=o
acc=0
for i in range(len(pred_outputs)):
    if pred_outputs[i]==act_outputs[i]:
        acc+=1
print(acc/len(pred_outputs))
```
